```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

import ast
```

```python
with open('../KeypointsDetector/hands.txt', 'r') as file:
    data = ast.literal_eval(file.read())

data[:5]
```

```python
keypoints = []
for i in range(0, len(data) - 2, 3):
    keypoints.append([data[i], data[i + 1], data[i + 2]])
```

```python
len(keypoints)
```

```python
#   
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#    keypoints
keypoints_3d = np.array(keypoints)

#   x, y, z
x = keypoints_3d[:, 0]
y = keypoints_3d[:, 1]
z = keypoints_3d[:, 2]

#   
ax.scatter(x, y, z, color='r', marker='o')

#  
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#  
plt.show()
```
